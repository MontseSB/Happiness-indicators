```python
#Visualize the generosity per country
fig,ax = plt.subplots(figsize = (20,6))
x= feli.Country_or_region
y= feli.Generosity
plt.plot(x, y, color= "orange")
plt.title("Generosity per country", fontsize=16)
plt.xlabel("Country", fontsize=14)
plt.ylabel("Generosity", fontsize=14)
```
